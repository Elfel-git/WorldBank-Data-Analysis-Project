# PHASE 0: DATA INTEGRATION
## Tích hợp các file CSV đầu vào thành dataset duy nhất

**Mục tiêu:**
- Load tất cả indicator CSV files từ folder `../nam`
- Hợp nhất chúng thành một DataFrame đơn nhất
- Chuẩn hóa cấu trúc: Countries (rows), Years (columns), Indicators (columns trong output)
- Lưu output: `dataset_merged.csv`

```python
import sys
import os
from pathlib import Path

# Add modules to path
sys.path.insert(0, str(Path.cwd() / 'modules'))

import pandas as pd
import numpy as np
import logging

from data_integration import DataIntegration

# Setup logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s'
)
logger = logging.getLogger(__name__)

print("✓ Modules imported successfully")
```

## Step 1: Load data từ folder `../nam`

```python
# Cấu hình input folder
input_folder = '../nam'

# Verify folder exists
if not os.path.exists(input_folder):
    raise FileNotFoundError(f"Input folder not found: {input_folder}")

csv_files = list(Path(input_folder).glob('*.csv'))
print(f"Found {len(csv_files)} CSV files in {input_folder}:")
for f in csv_files[:10]:  # Show first 10
    print(f"  - {f.name}")
if len(csv_files) > 10:
    print(f"  ... and {len(csv_files) - 10} more")
```

## Step 2: Initialize DataIntegration module

```python
# Create DataIntegration instance
integration = DataIntegration(input_folder=input_folder)

# Run the integration pipeline
merged_df = integration.run_integration()

print("\n" + "="*50)
print("INTEGRATION COMPLETE")
print("="*50)
```

## Step 3: Inspect merged data

```python
print("Shape:", merged_df.shape)
print("\nFirst few rows:")
print(merged_df.head())
print("\nColumn names:")
print(merged_df.columns.tolist())
print("\nData types:")
print(merged_df.dtypes)
```

```python
# Data Info
info = integration.get_data_info()
print("\n" + "="*50)
print("DATA SUMMARY")
print("="*50)
for key, value in info.items():
    if isinstance(value, (list, tuple)) and len(str(value)) > 100:
        print(f"{key}: {type(value).__name__} with {len(value)} items")
    else:
        print(f"{key}: {value}")

# Missing values
print("\n" + "="*50)
print("MISSING VALUES BY COLUMN")
print("="*50)
missing_pct = (merged_df.isna().sum() / len(merged_df) * 100).sort_values(ascending=False)
print(missing_pct[missing_pct > 0].head(10))
```

## Step 4: Save merged dataset

```python
# Create outputs folder if not exists
output_folder = Path('./outputs')
output_folder.mkdir(exist_ok=True)

# Save merged data
output_path = output_folder / 'dataset_merged.csv'
merged_df.to_csv(output_path, index=False, encoding='utf-8')

print(f"✓ Saved merged dataset to {output_path}")
print(f"  File size: {output_path.stat().st_size / 1024:.2f} KB")
print(f"  Shape: {merged_df.shape}")
```
