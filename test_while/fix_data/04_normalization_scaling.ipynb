# 04. Normalization & Scaling
## StandardScaler: achieve consistent scale across all features

**Objective**:
- Apply StandardScaler: (x - mean) / std
- Handle scale inconsistency (GDP: 0-174k vs Internet: 0-100)
- Prepare for ML algorithms (DTW, K-Means)

**Output**:
- Scaled dataset ready for clustering

```python
import pandas as pd
import numpy as np
from pathlib import Path
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')

# Load imputed data
data_dir = Path('./outputs')
consolidated_imputed = pd.read_csv(data_dir / 'consolidated_imputed.csv', index_col=0)

print(f"✓ Loaded: {consolidated_imputed.shape}")
```

## 1. Apply StandardScaler

```python
# StandardScaler: (x - mean) / std
print("\n" + "="*80)
print("STANDARDSCALER NORMALIZATION")
print("="*80 + "\n")

print("Before scaling:")
print(consolidated_imputed.describe().round(2))

# Apply StandardScaler
scaler = StandardScaler()
consolidated_scaled = pd.DataFrame(
    scaler.fit_transform(consolidated_imputed),
    columns=consolidated_imputed.columns,
    index=consolidated_imputed.index
)

print("\n" + "-"*80)
print("After scaling (StandardScaler):")
print(consolidated_scaled.describe().round(2))
```

## 2. Save Scaled Data & Scaler Object

```python
# Save scaled dataset
consolidated_scaled.to_csv(data_dir / 'consolidated_scaled.csv')

# Save scaler object for inverse transform later
import pickle
with open(data_dir / 'scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)

print(f"\n✓ Saved:")
print(f"  consolidated_scaled.csv")
print(f"  scaler.pkl (for inverse transform)")
print(f"\n→ Next: 05_ml_peer_clustering.ipynb (K-Means + DTW)")
```
