```python
import logging

# 1. Cấu hình logging
logging.basicConfig(
    filename='system.log',
    level=logging.DEBUG,
    format='%(asctime)s - %(levelname)s - %(message)s'
)
# Value error, exception, debug ....
# 1. Lỗi -> in màn hình 
# 2. Lỗi, ko đáng kể -> Warning 
# 3. -> Lưu thông tin -> .csv -> .log 

def thuc_hien_phep_chia(a, b):
    logging.info(f"INPUT: a={a}, b={b}") # Ghi lại đầu vào
    
    try:
        result = a / b
        logging.info(f"OUTPUT: Kết quả là {result}") # Ghi lại đầu ra thành công
        return result
    except ZeroDivisionError:
        logging.error("LỖI: Người dùng chia cho số 0") # Ghi lại lỗi
        return None

# Chạy thử
thuc_hien_phep_chia(10, 2)
thuc_hien_phep_chia(5, 0)
```
